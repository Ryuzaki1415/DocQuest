```python
from langchain_core.output_parsers import StrOutputParser

parser=StrOutputParser()
```

```python
from langchain_community.llms import Ollama
from langchain_community.embeddings import OllamaEmbeddings
model=Ollama(model="openhermes:latest")
embeddings=OllamaEmbeddings()
#model.invoke("tell me a joke")
```

```python
from langchain_community.document_loaders import PyPDFLoader

loader=PyPDFLoader("Lightweight_Single_Framework_for_Creating_Sparse_Generative_Pretrained_Transformers_of_Diverse_Applications_Using_Custom_Datasets.pdf")

pages=loader.load_and_split()
pages
```

```python
from langchain.prompts import PromptTemplate
template="""
Answer the Question  based on the context below.If you cannot answer the question, reply "IDK BRUH".

Context: {context}

Question: {question}
"""


prompt=PromptTemplate.from_template(template)
print(prompt.format(context="Here is some context",question="Here is a question"))
```

```python
from  langchain_community.vectorstores import DocArrayInMemorySearch

vectorstore=DocArrayInMemorySearch.from_documents(pages,embedding=embeddings)
```

```python
retriever=vectorstore.as_retriever()
retriever.invoke("SGPT")
```

```python
from operator import itemgetter    # Flow is like invoke(question)-->retriever -> context -> prompt -> model -> parser-> out 

chain=(
    {
        "context": itemgetter("question")|retriever,
        "question":itemgetter("question")
    }
    | prompt
    | model
    | parser
)
chain.invoke({"question":"what is talked about in the results section? What was the validation loss? "})

```
